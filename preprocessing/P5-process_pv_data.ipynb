## Feed-in Tariff Installations

```python
%%time 

df_fit_1 = pd.read_csv('../raw_data_to_edit/installation_report_apr2021_part_1.csv',skiprows=4)
df_fit_1.head()
```

```python
df_fit_2 = pd.read_csv('../raw_data_to_edit/installation_report_apr2021_part_2.csv',skiprows=4)
df_fit_2.head()
```

```python
df_fit_3 = pd.read_csv('../raw_data_to_edit/installation_report_apr2021_part_3.csv',skiprows=4)
df_fit_3.head()
```

```python
df_fit = pd.concat([df_fit_1, df_fit_2, df_fit_3])

#only keep PV technology
df_fit = df_fit[df_fit['Technology'] == 'Photovoltaic']
```

```python
#only keep domestic data
df_fit = df_fit[df_fit['Installation Type'] == 'Domestic']
```

We need to reduce the full dataset to give values on the LSOA regions.

```python
# make a new column of year/month data
```

```python
print(df_fit['Application date '].min())
print(df_fit['Application date '].max())
```

```python
df_fit['date_of_commission'] =  pd.to_datetime(df_fit['Commissioning date']).dt.to_period('M')
df_fit.head()
```

Create a pivot table of LSOA region by date counting the number of installations.

```python
lsoa_date_pv_capacity = pd.pivot_table(df_fit, values='Installed capacity', index=['LLSOA Code'],
                    columns=['date_of_commission'], aggfunc=np.sum, fill_value=0)
```

```python
lsoa_date_pv_count = pd.pivot_table(df_fit, values='Installed capacity', index=['LLSOA Code'],
                    columns=['date_of_commission'], aggfunc='count', fill_value=0)
```

```python
lsoa_date_pv_count.describe()
```

```python
lsoa_date_pv_count.head()
```

```python
lsoa_date_pv_count.reset_index(level=0, inplace=True)
lsoa_date_pv_count.columns
```

```python
lsoa_date_pv_count.rename(columns={'LLSOA Code': 'LSOA11CD'}, inplace=True)
```

```python
lsoa_codes_to_keep = pd.read_csv('../data/lsoa_codes_reduced.csv')
lsoa_codes_to_keep.head()
```

```python
df_fit_lsoa = pd.merge(
    lsoa_codes_to_keep,
    lsoa_date_pv_count,
    how="left",
    on='LSOA11CD',
)
```

```python
df_fit_lsoa.shape
```

```python
df_fit_lsoa.describe()
```

```python
mean_across_years = df_fit_lsoa.mean()
```
